#Phase 2: Preprocessing & Feature Engineering


```python
from sklearn.impute import KNNImputer

# ==========================
# CELL 1: MISSING VALUES + KNN IMPUTER
# ==========================

# 1) Summary of missing values BEFORE imputation
missing_summary = pd.DataFrame({
    "Missing": Df_diabetic.isnull().sum(),
    "Percentage (%)": (Df_diabetic.isnull().mean() * 100).round(2)
}).sort_values("Percentage (%)", ascending=False)

print("MISSING VALUES BEFORE IMPUTATION")
display(missing_summary.head(20))

# 2) Numeric columns to be imputed with KNN
numeric_cols = [
    "Age",
    "Diabetes duration (y)",
    "Height(cm)",
    "Weight(kg)",
    "BMI (kg/m2)",
    "SBP (mmHg)",
    "DBP (mmHg)",
    "HbA1c (%)",
    "FBG (mmol/L)",
    "TG (mmol/L)",
    "C-peptide (ng/mL)",
    "TC (mmol/L)",
    "HDLC (mmol/L)",
    "LDLC (mmol/L)"
]

# 3) Split features and target
target_col = "Diabetic nephropathy (DN)"
X = Df_diabetic.drop(columns=[target_col])
y = Df_diabetic[target_col]

# 4) Apply KNNImputer ONLY to numeric columns
imputer = KNNImputer(n_neighbors=5)

X_numeric_imputed = imputer.fit_transform(X[numeric_cols])
X_numeric_imputed = pd.DataFrame(
    X_numeric_imputed,
    columns=numeric_cols,
    index=X.index
)

# 5) Replace numeric columns in X with their imputed versions
X.update(X_numeric_imputed)

# 6) Check remaining NaNs in numeric columns AFTER imputation
missing_after = X[numeric_cols].isnull().sum()

print("\nMISSING VALUES IN NUMERIC COLUMNS AFTER KNN IMPUTATION:")
print(missing_after)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# ==========================
# CELL 2: FEATURE ENGINEERING + SCALING
# ==========================

# 1) Create engineered features based on clinical rationale
def add_engineered_features(df):
    df = df.copy()
    # Pulse pressure: hemodynamic load
    df["PP"] = df["SBP (mmHg)"] - df["DBP (mmHg)"]
    # Mean arterial pressure: average perfusion pressure
    df["MAP"] = (df["SBP (mmHg)"] + 2 * df["DBP (mmHg)"]) / 3
    # Non-HDL cholesterol: atherogenic lipoproteins
    df["Non_HDL"] = df["TC (mmol/L)"] - df["HDLC (mmol/L)"]
    # LDL / HDL ratio: balance between bad and protective cholesterol
    df["LDL_to_HDL"] = df["LDLC (mmol/L)"] / df["HDLC (mmol/L)"]
    # TyG index: surrogate of insulin resistance
    df["TyG"] = np.log(df["TG (mmol/L)"] * df["FBG (mmol/L)"] / 2)
    return df

# Apply feature engineering to X (already imputed in Cell 1)
X_fe = add_engineered_features(X)

print("Columns after feature engineering:")
print(X_fe.columns)

# 2) Select all numeric columns (original + engineered) for scaling
numeric_all = X_fe.select_dtypes(include=[np.number]).columns.tolist()

print("\nNumber of numeric features (original + engineered):", len(numeric_all))

# 3) Standard scaling
scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(X_fe[numeric_all])

X_scaled = pd.DataFrame(
    X_scaled_array,
    columns=numeric_all,
    index=X_fe.index
)

print("\nShape of X_scaled:", X_scaled.shape)
print("First 10 columns of X_scaled:")
print(X_scaled.columns[:10])
```

```python
"""
In the previous two cells, a structured preprocessing and feature engineering workflow was applied
to the original Diabetic Nephropathy dataset in order to prepare it for machine learning modeling.

First, we performed a careful inspection of the data quality. Obvious physiologically impossible
outliers (such as extremely high diastolic blood pressure and LDL values far beyond any realistic
clinical range) were identified and set to NaN so that they would not bias the analysis. Then we
quantified missing values across all variables and focused on the continuous clinical features
(age, diabetes duration, anthropometrics, blood pressure, glycemic markers, triglycerides, and
lipid profile), because these are the ones where numeric relationships between patients make sense
for distance‑based methods.

To handle missing data in these continuous variables, we applied KNNImputer with k=5 neighbors.
This method fills in missing entries by looking at similar patients in the feature space instead of
using simple mean or median values, which helps preserve the multivariate structure and realistic
clinical patterns in the dataset. After imputation, we verified that all numeric clinical
variables had zero missing values. At this point, we also separated the dataset into features (X)
and target (y), where the target corresponds to the binary label “Diabetic nephropathy (DN)”.

In the next cell, we implemented clinically motivated feature engineering, creating only derived
variables that are fully computable from the existing measurements and that have a clear
pathophysiological interpretation. Specifically, we added:
- Pulse Pressure (PP = SBP – DBP), capturing arterial stiffness and hemodynamic load.
- Mean Arterial Pressure (MAP = (SBP + 2*DBP) / 3), approximating average perfusion pressure.
- Non‑HDL cholesterol (Non_HDL = TC – HDLC), representing the pool of atherogenic lipoproteins.
- LDL/HDL ratio (LDL_to_HDL = LDLC / HDLC), reflecting the balance between “bad” and “protective”
  cholesterol.
- TyG index (TyG = ln(TG * FBG / 2)), a surrogate marker of insulin resistance based on fasting
  triglycerides and glucose.

These engineered features are widely used in cardiovascular and metabolic risk research and are
consistent with the clinical rationale and existing literature on diabetic complications, rather
than being arbitrary mathematical combinations.

After enriching the feature set with these new variables, we selected all numeric columns
(original continuous features plus the engineered ones) and applied standardization using
StandardScaler. This transformation rescales each numeric feature to have zero mean and unit
variance, which is important for many machine learning algorithms, especially those that are
sensitive to feature scales (such as distance‑based models and regularized linear models). The
result is a processed feature matrix (X_scaled) that is numerically well‑behaved and contains both
the original clinical information and a small set of clinically meaningful derived indices. This
matrix, together with the target vector y, is now ready to be used for train‑test splitting,
model training, and evaluation.
"""
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

# ==========================
# CELL: MULTICOLLINEARITY DIAGNOSTICS (READ‑ONLY)
# ==========================

# Work on a copy to avoid changing any existing object
X_diag = X_scaled.copy()

# 1) Correlation matrix between predictors (Pearson)
corr_matrix = X_diag.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0,
    square=True,
    cbar_kws={"shrink": 0.8}
)
plt.title("Correlation matrix of scaled predictors")
plt.show()

# 2) List highly correlated pairs (above a threshold)
corr_threshold = 0.90
high_corr_pairs = []

cols = corr_matrix.columns
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        r = corr_matrix.iloc[i, j]
        if abs(r) >= corr_threshold:
            high_corr_pairs.append((cols[i], cols[j], r))

print("\nHighly correlated pairs (|r| >= {:.2f}):".format(corr_threshold))
if not high_corr_pairs:
    print("No strongly correlated pairs found at this threshold.")
else:
    for v1, v2, r in high_corr_pairs:
        print(f"{v1}  <-->  {v2} : r = {r:.3f}")

# 3) Variance Inflation Factor (VIF) for each predictor
vif_data = pd.DataFrame()
vif_data["feature"] = X_diag.columns
vif_data["VIF"] = [
    variance_inflation_factor(X_diag.values, i)
    for i in range(X_diag.shape[1])
]

vif_data = vif_data.sort_values("VIF", ascending=False)

print("\nVariance Inflation Factor (VIF) for each feature:")
display(vif_data)

# 4) Simple rule‑of‑thumb guidance (no changes applied)
print(
    "\nInterpretation guideline (no features are removed here):\n"
    "- Correlation |r| >= 0.90 suggests very strong linear dependence between two predictors.\n"
    "- VIF > 5 (and especially > 10) indicates substantial multicollinearity.\n"
    "Use this table to decide whether it makes sense to apply feature selection methods\n"
    "such as LASSO or to manually drop one variable from highly redundant pairs."
)
```

```python
"""
Before building the predictive models, we explicitly assessed multicollinearity among all
predictors (original and engineered) using the correlation matrix and the Variance Inflation
Factor (VIF). The diagnostics showed very strong linear dependencies in several blocks of
variables. In particular, there was a high correlation between DBP and MAP, as well as between
TC and Non_HDL, and multiple predictors (SBP, DBP, PP, MAP, TC, HDLC, Non_HDL) exhibited
extremely high or even infinite VIF values. This indicates that many of the engineered features
(PP, MAP, Non_HDL, LDL_to_HDL) are almost linear combinations of the raw clinical measures
(SBP, DBP, TC, HDLC, LDLC).

From a clinical point of view, these engineered indices can still be valuable: they summarize
hemodynamic load (PP, MAP) or atherogenic lipid burden (Non_HDL, LDL_to_HDL) and are widely
used in cardiovascular and nephropathy research. However, their strong collinearity with the
underlying raw variables means that, in strictly linear models, coefficient estimates may become
unstable and hard to interpret. For this reason, future work could explore simplified feature
sets (e.g., keeping only one representative variable per highly correlated block) or apply
regularization-based feature selection (such as LASSO) to automatically decide which of these
indices to retain.

In the present study, we proceed to train machine learning models using the full set of scaled
features, including both original and engineered predictors. Tree-based ensembles (Random Forest
and XGBoost), Support Vector Machines with non-linear kernels, and neural networks are generally
more robust to multicollinearity than classical linear regression models, because they do not rely
on interpreting individual coefficients and can handle redundant predictors without violating key
assumptions. Therefore, the presence of highly correlated features is not expected to critically
affect the performance of these models, although it may be revisited in future investigations if
model simplification or interpretability become a priority.
"""
```

#Phase 3: Generation of models

```python
from sklearn.model_selection import train_test_split

# ==========================
# TRAIN–TEST SPLIT
# ==========================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,          # feature matrix (scaled)
    y,                 # target vector (DN label)
    test_size=0.30,    # 30% for test, 70% for train
    stratify=y,        # preserves class proportion in both sets
    random_state=42    # reproducible split
)

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape:", X_test.shape, y_test.shape)
```
